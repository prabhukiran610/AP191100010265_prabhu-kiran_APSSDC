- Take any data set and plot different kind of graphs

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/LavanyaPolamarasetty/Datasets/master/Classification/Social_Network_Ads.csv")
df.head()
```

```python
# Line plot

plt.plot(df["Age"])
plt.title("Line Plot")
plt.xlabel("Age")
plt.show()
```

```python
# Scatter plot

plt.scatter(df["Age"],df["EstimatedSalary"])
plt.title("Scatter Plot")
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.show()
```

```python
# Bar graph


plt.bar(df["Gender"],df["EstimatedSalary"])
plt.title("Bar graph")
plt.xlabel("Gender")
plt.ylabel("Estimated Salary")
plt.show()
```

```python
# Box plot

plt.boxplot(df["EstimatedSalary"])
plt.title("Box Plot")
plt.xlabel("Estimated Salary")
plt.show()
```

```python
# Histogram

plt.hist(df["Age"])
plt.title("Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency count of age")
```

```python
# Pie Chart

plt.pie(df["EstimatedSalary"][:10],labels = df["Age"][:10], autopct = "% 1.1f%%")
plt.show()
```
